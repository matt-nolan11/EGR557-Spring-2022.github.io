## Kinetic Energy Analysis:

```python
# Create Variables for Gait Size and Total Kinetic Energy

Gait_Size = numpy.array([4/7, 1/2, 1/3, 1/4])
Total_Kinetic_Energy = numpy.array([0.03686251840505446, 0.038137493868378246, 0.03105457942700576, 0.02908399879372533])
```

```python
plt.plot(Gait_Size, Total_Kinetic_Energy)
```

```python
# Define Basis Functions for Creating the Optimization Function

x = numpy.r_[0.25:0.580:.005]
A = numpy.array([(Gait_Size**2),(Gait_Size)]).T
```

```python
def myfunc(k):
    # make sure our coefficients are in the form of a numpy array
    k = numpy.array(k)
    # generate y* = Ak^T
    y_model = A.dot(k.T)
    # sum the square of the error of our model against the input data, y_rand
    error = ((y_model-Total_Kinetic_Energy)**2).sum()
    #return the error
    return error
```

```python
# Define Initial Guess for Coefficients:

ini = [0.1]*A.shape[1]
ini
```

```python
# Run the Optimization Functions and Find Corresponding Coefficients:

sol = scipy.optimize.minimize(myfunc,ini)
k_optimum = sol.x
k_optimum
```

```python
# Plot Resulting Equation to the Data Found

A = numpy.array([(x**2),(x)]).T
y_model = A.dot(k_optimum.T)
fig = plt.figure()
ax = fig.add_subplot()
a = ax.plot(Gait_Size,Total_Kinetic_Energy,'.')
b = ax.plot(x,y_model)
ax.legend(a+b,['data','model'])
```

```python
# Deduce Gait Size for Maximum Kinetic Energy:

y_model = y_model.tolist()
Max_Kinetic_Energy = max(y_model)
print('Maximum Kinetic Energy: ')
print(Max_Kinetic_Energy)
max_index = y_model.index(Max_Kinetic_Energy)
Max_Gait_Size = x[max_index]
print('Gait Size for Maximum Kinetic Energy')
print(Max_Gait_Size)
```
